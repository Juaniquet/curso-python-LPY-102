### ðŸ§© Problema B â€” Sistema lineal

## 6ï¸âƒ£ Ejercicio integrador â€” Dataset numÃ©rico

Dado un dataset simulado (puntuaciones por examen):
```python
notas = np.array([
    [7,8,9],
    [5,6,7],
    [9,9,10],
    [4,5,6]
])
```
### ðŸ§© Objetivo
1. Calcula la media por alumno
2. Calcula la media por examen
3. ObtÃ©n una matriz normalizada restando la media por examen
4. Calcula el vector de varianzas por examen
5. Extrae los valores mayores a 8 usando boolean indexing

Escribe aquÃ­ tu trabajo:

````python
notas = np.array([
    [7,8,9],
    [5,6,7],
    [9,9,10],
    [4,5,6]
])

# Tu soluciÃ³n aquÃ­
media_alumno = notas.mean(axis=1)
media_examen = notas.mean(axis=0)
normal = notas - media_examen
vari = notas.var(axis=0)
mas8= notas[notas>8]

media_alumno, media_examen, normal, vari, mas8
````
